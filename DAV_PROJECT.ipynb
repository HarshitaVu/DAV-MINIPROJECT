```python
def convert_k(value):
    if isinstance(value, str):
        value = value.replace(',', '').strip()
        if value.endswith('k'):
            return float(value[:-1]) * 1000
        elif value.isnumeric():
            return float(value)
        else:
            try:
                return float(value)
            except:
                return np.nan
    return value

numeric_cols = ['star', 'fork', 'watch', 'issue', 'pull_requests', 'commits', 'branches', 'packages', 'releases', 'contributers']
for col in numeric_cols:
    df[col] = df[col].apply(convert_k)
```

```python
stars_array = df['star'].to_numpy()
print("Stars - Max:", np.max(stars_array))
print("Stars - Min:", np.min(stars_array))
print("Stars - Median:", np.median(stars_array))
print("Stars - 25th Percentile:", np.percentile(stars_array, 25))
print("Stars - 75th Percentile:", np.percentile(stars_array, 75))
```

```python
forks = df['fork'].to_numpy()
high_fork_mask = forks > 1000
print("Repositories with >1000 forks:")
print(df[high_fork_mask][['name', 'fork']])
```

```python
stars = df['star'].to_numpy()
watches = df['watch'].to_numpy()
correlation = np.corrcoef(stars, watches)
print("Correlation between Stars and Watches:\n", correlation)
```

```python
normalized_stars = (stars_array - np.min(stars_array)) / (np.max(stars_array) - np.min(stars_array))
print("Normalized Stars (first 5):", normalized_stars[:5])
```

```python
forks_array = df['fork'].to_numpy()
combined_score = np.multiply(stars_array, forks_array)
print("Combined Score (Stars × Forks) for first 5 repos:", combined_score[:5])
```

```python
contributors_array = df['contributers'].to_numpy()
zero_contribs = np.sum(contributors_array == 0)
print("Number of Repos with Zero Contributors:", zero_contribs)

```

```python
z_contributors = (contributors_array - np.mean(contributors_array)) / np.std(contributors_array)
print("Z-score Standardized Contributors (first 5):", z_contributors[:5])
```
